# Putting all together

```python
def get_url(jobRole):
    url = searchWith_jobTitle.format(jobRole)
    return url

def get_record(job_title):
    """ Get Job Data From Single Record"""
    atag = job_title.a
    job_title_name = atag.text
    job_title_url = main_url+ atag.get('href')
    
    company_atag = job_title.find('p','search__job-subtitle').a
    company_name = company_atag.text.strip()
    
    job_addition_row = job_title.find('div','search__job-additional')
    
    location_stag = job_addition_row.find('p','search__job-location').span
    location = location_stag.text.strip()
    
    salary_stag = job_addition_row.span
    salary = salary_stag.text.strip()
    

    job_footer_rows = job_title.find_all('div','search__job-footer')
    job_footer_row = job_footer_rows[0]
    job_date_tag = job_footer_row.find('p','search__job-posted').span
    job_date = job_date_tag.text.strip()
    
    today = datetime.today().strftime('%Y-%m-%d')
    
    benefit = job_title.find('p','benefit').text.strip()
    benefit = benefit[10:]
    
    highlight = job_title.find('p','highlights').text.strip()
    highlight = highlight[11:]
    
    career_opt = job_title.find('p','').text
    career_opt = career_opt[22:]
    
    record = (job_title_name,company_name,location,job_date,benefit,highlight,career_opt,salary,job_title_url)
    
    return record



def main(jobRole):
    """Run main program """
    records = []
    url = get_url(jobRole)
    response = requests.get(url)
    bf_soup = BeautifulSoup(response.text,'html.parser')
    job_titles = bf_soup.find_all('div','search__job')
    for job_title in job_titles:
        record = get_record(job_title)
        records.append(record)
        df = pd.DataFrame(records)
        df.columns = ['JobTitle', 'Company', 'Location', 'PostedDate', 'Benefit','HightLight','Career Opportunities', 'Salary','JobURL']
    return df
```

# Get Final data From Web

```python
# Run main function here
#Data Analyst
dataAnalyst_df = main('data  analyst')
dataAnalyst_df
```

```python
#save job data as csv
dataAnalyst_df.to_csv('dataAnalystJob.csv',encoding='utf-8')
```
